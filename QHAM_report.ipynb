```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def ratio(datafile):
    datafile['m_n'] = datafile['n_mems']/datafile['n_qubits']
    datafile.sort_values(by=['n_updates', 'm_n'])
    return datafile
```

```python
df4 = pd.read_csv('/home/rajarsi/Documents/QuantumML.project/QHAM.report/EXP_ n=[8]_ m=[4, 6, 8, 12, 14, 16]')
df5 = pd.read_csv('/home/rajarsi/Documents/QuantumML.project/QHAM.report/EXP_ n=[5]_ m=[3, 4, 5, 6, 8, 10]')
df8 = pd.read_csv('/home/rajarsi/Documents/QuantumML.project/QHAM.report/EXP_ n=[8]_ m=[4, 6, 8, 12, 14, 16]')
df10 = pd.read_csv('/home/rajarsi/Documents/QuantumML.project/QHAM.report/EXP_ n=[10]_ m=[4, 8, 12, 14, 16]')
df12 = pd.read_csv('/home/rajarsi/Documents/QuantumML.project/QHAM.report/EXP_ n=[12]_ m=[4, 12, 18, 24, 28, 32]')
```

```python
frames = [df4,df5,df8,df10,df12]
frames = [ratio(df) for df in frames]
df = pd.concat(frames)
# df
```
